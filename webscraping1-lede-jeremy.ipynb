```python
cdc_html = requests.get("https://www.health.ny.gov/diseases/communicable/zoonoses/mpox/data/").text

print(cdc_html)
```

```python
#2. To get to scraping, we're going to transform HTML into a DOM 
#Two libraries to do this, BeautifulSoup and lxml. We're focusing on BeautifulSoup bc a bit more popular. But both good & pros and cons.

!pip install beautifulsoup4
```

```python
from bs4 import BeautifulSoup
```

```python
#3. Converting HTML into BeautifulSoup object
#variable (can call it anything)

cdc_soup = BeautifulSoup(cdc_html) 
```

```python
type(cdc_soup)
```

```python
#now we're operating on the DOM not the HTML

cdc_soup.title
```

```python
#can also do it like this (can drill down through head to get title) - but

cdc_soup.head.title
```

```python
#How to get p tag? (this notation is only going to give you the first matching thing, not the subsequent matching thing)
cdc_soup.body.p
```

```python
#Another way to get all p tags (or all occurrences of some element) is to do "find_all"
#**find_all is specific to tags 

all_p_tags = cdc_soup.body.find_all("p")
all_p_tags
```
